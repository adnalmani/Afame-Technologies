```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset .csv')
```

```python
cuisine_counts = df['Cuisines'].value_counts()
top_three_cuisines = cuisine_counts.head(3)
top_three_df = top_three_cuisines.reset_index()
top_three_df.columns = ['Cuisine', 'Count']
plt.figure(figsize=(10, 6))
sns.barplot(x='Cuisine', y='Count', data=top_three_df, palette='viridis')
plt.title('Top 3 Most Common Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Number of Occurrences')
plt.show()
```

```python
cuisine_counts = df['Cuisines'].value_counts()
top_three_cuisines = cuisine_counts.head(3)
total_restaurants = len(df)
top_three_percentages = (top_three_cuisines / total_restaurants) * 100
labels = top_three_percentages.index
sizes = top_three_percentages.values
colors = ['#ff9999','#66b3ff','#99ff99']  # Color palette for pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})
plt.title('Percentage of Restaurants Serving Each of the Top 3 Cuisines')
plt.show()
```
